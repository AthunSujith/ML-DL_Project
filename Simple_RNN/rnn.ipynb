```python
#import
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical

```

```python
text = "hello world"
chars = sorted(list(set(text)))
char_to_idx = {c:i for i,c in enumerate(chars)}
idx_to_char = {i:c for i,c in enumerate(chars)}
```

```python
seq_length = 3
X, y = [], []

for i in range(len(text) - seq_length):
    seq = text[i:i+seq_length]       # input sequence like "hel"
    target = text[i+seq_length]      # next character like "l"
    X.append([char_to_idx[ch] for ch in seq])
    y.append(char_to_idx[target])
```

```python
X = to_categorical(X, num_classes=len(chars))
y = to_categorical(y, num_classes=len(chars))
```

```python
model = Sequential([
    SimpleRNN(32, input_shape=(seq_length, len(chars))),
    Dense(len(chars), activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=100)
```

```python
def predict_next_char(input_seq):
    x = np.array([[char_to_idx[c] for c in input_seq]])
    x = to_categorical(x, num_classes=len(chars))
    pred = model.predict(x)[0]
    return idx_to_char[np.argmax(pred)]
```

```python
print(predict_next_char("hel"))  # Output might be 'l' or 'o'
```
